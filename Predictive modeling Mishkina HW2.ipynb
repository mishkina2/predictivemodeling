```python
from statsmodels.tsa.stattools import adfuller
```

```python
ticker= "tsla"
tsla_data = yf.download(ticker, start="2023-01-18", end="2024-01-17")
tsla_data = pdr.get_data_yahoo('TSLA', start='2023-01-18', end='2024-01-17')
```

```python
tsla_data.head(10)
```

```python
print (tsla_data)
```

```python
tsla_data.head()
```

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(tsla_data['Close'].values, autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

Apply Log transformation and test again

```python
from numpy import log

result = adfuller(log(tsla_data['Close'].values), autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

After applying log transformaion data is stationary

```python
pip install arch
```

```python
tsla_data['Return'] = 100 * (tsla_data['Close'].pct_change())
tsla_data.dropna(inplace=True)
```

```python
from arch import arch_model
from arch.__future__ import reindexing

garch_model = arch_model(tsla_data['Return'], p = 1, q = 1,
                      mean = 'constant', vol = 'GARCH', dist = 'normal')

gm_result = garch_model.fit(disp='off')
print(gm_result.params)

print('\n')

gm_forecast = gm_result.forecast(horizon = 5)
print(gm_forecast.variance[-1:])
```

```python
garch_result = garch_model.fit()

print(garch_result.summary())
```

```python
garch_result.plot()
plt.show()
```

```python
garch_t_model = arch_model(tsla_data['Return'], p = 1, q = 1, mean = 'constant', vol = 'GARCH', dist = 'skewt')
garch_t_result = garch_t_model.fit()
garch_t_result.summary()
```

```python
garch_gjr_model = arch_model(tsla_data['Return'], p = 1, o = 1, q = 1, mean = 'constant', vol = 'GARCH', dist = 'normal')
garch_gjr_result = garch_gjr_model.fit()
garch_gjr_result.summary()
```

```python
egarch_model = arch_model(tsla_data['Return'], p = 1, o = 1, q = 1, mean = 'constant', vol = 'EGARCH', dist = 'normal')
egarch_result = egarch_model.fit()
egarch_result.summary()
```
